```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./latency.csv')



sort_df = df.sort_values('rpc_rate (rpc/sec)').iloc[:500]

display(sort_df[])
```

```python
for column in sort_df.columns[:-1]:  # Exclude the 'rpc_rate' column
    if column in ['Max latency (ns)', 'moving avg (ns)', 'p999 latency (ns)']:
        continue
    plt.scatter(sort_df['rpc_rate (rpc/sec)'], sort_df[column], label=column)

plt.title('Columns vs rpc_rate')
plt.xlabel('rpc_rate (rpc/sec)')
plt.ylabel('Latency in nanoseconds')
plt.legend()
plt.grid(True)
plt.show()
```
